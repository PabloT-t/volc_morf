```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

##cargar datos
ruta = r'C:\Users\56931\Documents\MORVOLCFINAL.xlsx'  # asegúrate que esté en el mismo directorio o da la ruta completa
df = pd.read_excel(ruta)

##para que tome valores de la fila 0 como titulo y a la vez elimine esa fila de los datos y se reordena
df.columns = df.iloc[0]       
df = df.drop(index=0)         
df = df.reset_index(drop=True)

##seleccion de columnas para correlacion
columnas = ['Wb(m)', 'H(m)', 'Vol(m3)', 'I_Ero', 'VolEro(m3)', 'Pend(º)', 'H/Wb', '1/Elipt_Ba']
df_corr = df[columnas].apply(pd.to_numeric, errors='coerce') 

##calculo matriz de correlacion
matriz_corr = df_corr.corr(method='pearson')

##plot
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_corr, annot=True, cmap='coolwarm', fmt=".2f", square=True, linewidths=0.5)
plt.title('Matriz de Correlación de Parámetros Morfométricos')
plt.tight_layout()

##se guarda primero por que al llamar .show la imagen se quedaba vacia; esto es opcional por si se quiere guardar en buena calidad; sino ##
ruta_imagen = r"C:\Users\56931\Documents\matriz_correlacion.png"
plt.savefig(ruta_imagen, dpi=300, bbox_inches='tight')

##
plt.show()


```
